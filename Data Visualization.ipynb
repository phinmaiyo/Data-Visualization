# plot a line chart for oranges and apples in one graph.Add labels and markerfacecolours to both line charts.A line chart for oranges should be dotted also add a horizontal line where y=0.93 and label it as median yield and vertical line wher year is 2003 and 2008.

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896 ]
```

```python
plt.plot(years,apples,'--',color='g',label='apples',marker='o',markerfacecolor='r')
plt.plot(years,oranges,'--',color='orange',label='oranges',marker='p',markerfacecolor='k')
plt.xlabel('Year')
plt.ylabel('Yield(tons per hectare)')
plt.axvline(x=2003,linestyle='--',c='silver')
plt.axvline(x=2008,c='m')
plt.axhline(y=0.93,linestyle='--',c='b',label='median yield')
plt.legend();
```
